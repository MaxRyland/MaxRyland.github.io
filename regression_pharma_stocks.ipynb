```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
sns.set()
```

```python
stocks=pd.read_csv('pharma_stocks.csv')
```

```python
x = stocks['Date'].values
x = pd.to_datetime(x) #let's make this a proper date
y1 = stocks[stocks.columns[1]]
y2 = stocks[stocks.columns[2]]
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(stocks.columns[1:], ncol=2, loc='upper left');
```

```python
myregression = sns.regplot(x='JNJ',y='MRNA',data=stocks,line_kws={"color":"r"})
```

```python
from scipy.stats import linregress
```

```python
slope, intercept, rvalue, pvalue, stderr = linregress(x=stocks.dropna()['JNJ'], y=stocks.dropna()['MRNA'])
```

```python
print('y = ',np.round(slope,2),' * x + ',np.round(intercept,2))
print('R = ',np.round(rvalue**2,2))
```

```python
np.corrcoef(stocks.dropna()['JNJ'], stocks.dropna()['MRNA'])
```
